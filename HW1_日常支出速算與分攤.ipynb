```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1ge-RkD_jfq4NNu5egU2Fk48pBttx6vpIvsPWIh96B_o/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('工作表1').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python
# 讓使用者輸入資料
while True:
    date_str = input("請輸入日期 (格式：YYYY-MM-DD，留空則有效): ")
    if not date_str:
        break
    try:
        datetime.datetime.strptime(date_str, '%Y-%m-%d')
        break
    except ValueError:
        print("日期格式不正確，請重新輸入 (YYYY-MM-DD)。")

while True:
    time_str = input("請輸入時間 (格式：HH:MM，留空則有效): ")
    if not time_str:
        break
    try:
        datetime.datetime.strptime(time_str, '%H:%M')
        break
    except ValueError:
        print("時間格式不正確，請重新輸入 (HH:MM)。")

categories = ["食物", "飲料", "交通", "書", "其他"]
for idx, category in enumerate(categories):
    print(f"{idx + 1}. {category}")
category_idx = int(input("請輸入分類(編號): "))


item = input("請輸入品項: ")
amount = int(input("請輸入金額: "))
```

```python
# 使用 mask 結合 isna() 將 NaN 值替換為 None，以便 gspread 處理
df_cleaned = df.mask(df.isna(), None)

# 將 DataFrame 轉換成列表的列表 (list of lists)，這是 gspread 支援的資料格式
data_to_write = df_cleaned.values.tolist()

# 第一步：取得工作表物件
worksheet = gsheets.worksheet('工作表1')

# 第二步：將資料寫入工作表
worksheet.append_rows(values=data_to_write, value_input_option='USER_ENTERED')
print("資料已成功寫入 Google 工作表！")

df = pd.DataFrame(columns=df.columns)
print("DataFrame 已清空。")
```
